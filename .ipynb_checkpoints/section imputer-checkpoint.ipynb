```python
import pandas as pd
import operator
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("train.csv")
```

```python
def get_columns(df=None):
    return df.columns
```

```python
def get_na_count(df=None):
    count_col_nulls = []
    for col in train_df.columns:
        count_col_nulls.append((col, train_df[col].isna().sum()))
    count_col_nulls = sorted(count_col_nulls, key=operator.itemgetter(1), reverse=True)
    return count_col_nulls
```

```python
get_na_count(train_df)
```

```python
train_df.drop(["MiscFeature", "FireplaceQu", "Alley","PoolQC", "Fence"], inplace=True, axis=1)
```

```python
get_na_count(train_df)
```

```python
col_with_na = [col_data[0] for col_data in get_na_count(train_df) if train_df[col_data[0]].dtypes=="O" and col_data[1]>0]
```

```python
train_df["TotalBaths"] = (0.5*train_df["BsmtHalfBath"])+(train_df["BsmtFullBath"])+(0.5*train_df["HalfBath"])+(train_df["FullBath"])
```

```python
train_df["BasmentBaths"] = (0.5*train_df["BsmtHalfBath"])+(train_df["BsmtFullBath"])
```

```python
x_train = train_df[result_col]
```

```python
y_train = train_df["SalePrice"]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mscl = MinMaxScaler()
```

```python
for col in x_train.columns:
    x_train[col] = mscl.fit_transform(x_train[[col]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=45)
```

```python
x_train.shape, y_train.shape, x_val.shape, y_val.shape
```

```python
test_df = pd.read_csv("test.csv")
```

```python
test_df["Yr_built_remod_diff"] = test_df["YearBuilt"] - test_df["YearRemodAdd"]
test_df["Yr_sold_remod_diff"] = abs(test_df["YearRemodAdd"] - test_df["YrSold"])
test_df["TotalBaths"] = (0.5*test_df["BsmtHalfBath"])+(test_df["BsmtFullBath"])+(0.5*test_df["HalfBath"])+(test_df["FullBath"])
test_df["BasmentBaths"] = (0.5*test_df["BsmtHalfBath"])+(test_df["BsmtFullBath"])
test_df["RoomBaths"] = (0.5*test_df["HalfBath"])+(test_df["FullBath"])
test_df["TotalPorchArea"] = test_df["OpenPorchSF"]+test_df["EnclosedPorch"] + test_df["3SsnPorch"] + test_df["ScreenPorch"]
```

```python
x_test = test_df[result_col]
```

```python
x_test.isna().sum()
```

```python
import numpy as np
x_test["BsmtQual"] = x_test["BsmtQual"].fillna("None")
x_test["KitchenQual"] = x_test["KitchenQual"].fillna("None")
x_test["TotalBsmtSF"] = x_test["TotalBsmtSF"].fillna(np.mean(x_test["TotalBsmtSF"]))
x_test["GarageCars"] = x_test["GarageCars"].fillna(np.mean(x_test["GarageCars"]))
x_test["GarageArea"] = x_test["GarageArea"].fillna(np.mean(x_test["GarageArea"]))
x_test["TotalBaths"] = x_test["TotalBaths"].fillna(np.mean(x_test["TotalBaths"]))
```

```python
for col in object_cols:
    x_test[col] = le.fit_transform(x_test[col])
```

```python
x_test
```

```python
for col in x_test.columns:
    x_test[col] = mscl.fit_transform(x_test[[col]])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
model_rf = RandomForestRegressor(n_estimators=1000)
model_rf.fit(x_train, y_train)
model_rf.score(x_val, y_val)
result = model_rf.predict(x_test)
result
sub = pd.read_csv("sample_submission.csv")
sub["SalePrice"] = result
sub.to_csv("section_imputer_rf.csv", index=False)
```

```python
model_rf.score(x_val, y_val)
```

```python
from sklearn.metrics import make_scorer
def rmse(y_true, y_pred):
    return np.sqrt(((y_true-y_pred)**2).mean())
param_grid = {
    "n_estimators": [i for i in range(100,300,10)],
    "bootstrap": [True],
}
gscv = GridSearchCV(estimator=model_rf,
                    param_grid=param_grid,
                    n_jobs=-1,
                    verbose=2,
                    scoring=make_scorer(rmse, greater_is_better=False))
```

```python
gscv.fit(x_train, y_train)
```

```python
gscv.best_params_
```

```python
gscv.score(x_val, y_val)
```

```python
gscv.get_params()
```

```python
result = gscv.predict(x_test)
sub = pd.read_csv("sample_submission.csv")
sub["SalePrice"] = result
sub.to_csv("section_imputer_rf_gscv_2.csv", index=False)
```
